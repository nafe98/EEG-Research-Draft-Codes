# Importing Libraries 

```python
import scipy.io
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import welch
```

# Load the dataset

```python
# Load the dataset
file_path = 'EEG driver drowsiness dataset.mat'
data = scipy.io.loadmat(file_path)

# Explore the structure of the dataset
data.keys()
```

# The dimensions of the data components

```python
# Examining the dimensions and format of EEGsample, subindex, and substate
eeg_samples = data['EEGsample']
subindex = data['subindex']
substate = data['substate']

eeg_samples.shape, subindex.shape, substate.shape
```

# Function to plot EEG sample

```python
def plot_eeg_sample(eeg_data, sample_rate, title):
    plt.figure(figsize=(15, 6))
    for i in range(eeg_data.shape[0]):
        plt.plot(np.arange(eeg_data.shape[1]) / sample_rate, eeg_data[i, :] + i*100, label=f'Channel {i+1}')
    plt.title(title)
    plt.xlabel('Time (s)')
    plt.ylabel('Amplitude')
    plt.legend()
    plt.show()
```

# Function to plot Power Spectral Density (PSD)

```python
def plot_psd(eeg_data, sample_rate, title):
    plt.figure(figsize=(15, 6))
    freqs, psd = welch(eeg_data, fs=sample_rate, nperseg=512)
    for i in range(eeg_data.shape[0]):
        plt.semilogy(freqs, psd[i, :], label=f'Channel {i+1}')
    plt.title(title)
    plt.xlabel('Frequency (Hz)')
    plt.ylabel('Power/Frequency (dB/Hz)')
    plt.legend()
    plt.show()
```

# Selecting samples for Alert and Drowsy states

```python
alert_sample = eeg_samples[substate.ravel() == 0][0]  # First sample of alert state
drowsy_sample = eeg_samples[substate.ravel() == 1][0]  # First sample of drowsy state
sample_rate = 128  # Sampling rate
```

# Plotting time-domain and frequency-domain representations

```python
plot_eeg_sample(alert_sample, sample_rate, 'EEG Sample - Alert State')
```

```python
plot_eeg_sample(drowsy_sample, sample_rate, 'EEG Sample - Drowsy State')
```

```python
plot_psd(alert_sample, sample_rate, 'Power Spectral Density - Alert State')
```

```python
plot_psd(drowsy_sample, sample_rate, 'Power Spectral Density - Drowsy State')
```

```python
def compute_average_psd(eeg_data, sample_rate):
    """Compute average Power Spectral Density (PSD) for a set of EEG samples."""
    psd_avg = np.zeros((eeg_data.shape[1], int(384/2+1)))  # PSD shape for each channel
    for sample in eeg_data:
        freqs, psd = welch(sample, fs=sample_rate, nperseg=384)
        psd_avg += psd
    psd_avg /= len(eeg_data)
    return freqs, psd_avg

# Separate the EEG samples into alert and drowsy states
alert_eeg_samples = eeg_samples[substate.ravel() == 0]
drowsy_eeg_samples = eeg_samples[substate.ravel() == 1]

# Compute average PSD for Alert and Drowsy states
freqs, avg_psd_alert = compute_average_psd(alert_eeg_samples, sample_rate)
_, avg_psd_drowsy = compute_average_psd(drowsy_eeg_samples, sample_rate)

# Plotting the average PSDs
plt.figure(figsize=(15, 6))
for i in range(avg_psd_alert.shape[0]):
    plt.semilogy(freqs, avg_psd_alert[i, :], label=f'Channel {i+1} - Alert')
    plt.semilogy(freqs, avg_psd_drowsy[i, :], '--', label=f'Channel {i+1} - Drowsy')
plt.title('Average Power Spectral Density - Alert vs Drowsy States')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Power/Frequency (dB/Hz)')
plt.legend()
plt.show()
```

```python
from scipy.stats import ttest_ind

# Frequency bands
bands = {
    'Delta': (0.5, 4),
    'Theta': (4, 8),
    'Alpha': (8, 12),
    'Beta': (12, 30),
    'Gamma': (30, 50)
}

# Function to calculate average power in a frequency band
def band_power(psd, freqs, band):
    idx_band = np.logical_and(freqs >= band[0], freqs <= band[1])
    return psd[:, idx_band].mean(axis=1)

# Calculate and compare the band powers for alert and drowsy states
band_powers_results = {}
for band_name, band_range in bands.items():
    alert_band_power = band_power(avg_psd_alert, freqs, band_range)
    drowsy_band_power = band_power(avg_psd_drowsy, freqs, band_range)
    
    # Perform t-test
    t_stat, p_value = ttest_ind(alert_band_power, drowsy_band_power)
    band_powers_results[band_name] = (t_stat, p_value)

band_powers_results

```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Function to extract features (average power in frequency bands) for each EEG sample
def extract_features(eeg_dataset, freqs):
    features = np.zeros((eeg_dataset.shape[0], len(bands)))  # Each sample will have features for each band
    for i, sample in enumerate(eeg_dataset):
        _, psd = welch(sample, fs=sample_rate, nperseg=384)
        for j, (band_name, band_range) in enumerate(bands.items()):
            features[i, j] = band_power(psd, freqs, band_range).mean()
    return features

# Extracting features for the whole dataset
eeg_features = extract_features(eeg_samples, freqs)

# Preparing the labels (substate)
labels = substate.ravel()

# Splitting the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(eeg_features, labels, test_size=0.2, random_state=42)

# Feature scaling
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.svm import SVC
from sklearn.metrics import classification_report, accuracy_score

# Training the SVM classifier
svm_model = SVC(kernel='rbf', random_state=42)
svm_model.fit(X_train, y_train)

# Predicting on the test set
y_pred = svm_model.predict(X_test)

# Evaluating the model
accuracy = accuracy_score(y_test, y_pred)
classification_report_str = classification_report(y_test, y_pred, target_names=['Alert', 'Drowsy'])

accuracy, classification_report_str
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.utils.class_weight import compute_class_weight

# Computing class weights to address potential class imbalance
class_weights = compute_class_weight('balanced', classes=np.unique(y_train), y=y_train)
class_weights_dict = {0: class_weights[0], 1: class_weights[1]}

# Setting up the parameter grid for grid search
param_grid = {
    'C': [0.1, 1, 10, 100],  # Regularization parameter
    'gamma': ['scale', 'auto'],  # Kernel coefficient
    'kernel': ['rbf', 'poly', 'sigmoid']  # Type of SVM kernel
}

# Creating a GridSearchCV object
svm_grid = GridSearchCV(SVC(class_weight=class_weights_dict, random_state=42), 
                        param_grid, 
                        cv=3, 
                        verbose=2, 
                        scoring='accuracy')

# Performing grid search
svm_grid.fit(X_train, y_train)

# Best parameters and best score
best_params = svm_grid.best_params_
best_score = svm_grid.best_score_

best_params, best_score
```

```python
# Retraining the SVM model with the best parameters
optimized_svm_model = SVC(C=100, gamma='scale', kernel='rbf', class_weight=class_weights_dict, random_state=42)
optimized_svm_model.fit(X_train, y_train)

# Predicting on the test set with the optimized model
y_pred_optimized = optimized_svm_model.predict(X_test)

# Evaluating the optimized model
optimized_accuracy = accuracy_score(y_test, y_pred_optimized)
optimized_classification_report = classification_report(y_test, y_pred_optimized, target_names=['Alert', 'Drowsy'])

optimized_accuracy, optimized_classification_report
```

```python
import pywt

# Function to extract time-frequency features using Wavelet Transform
def extract_time_frequency_features(eeg_dataset, wavelet='db4'):
    # Number of features: For each frequency band in each channel, we'll have mean and variance
    n_features = len(bands) * 2 * eeg_dataset.shape[1]  # Two features (mean, variance) for each band in each channel
    features = np.zeros((eeg_dataset.shape[0], n_features))
    
    for i, sample in enumerate(eeg_dataset):
        feature_vector = []
        for channel_data in sample:
            # Applying Wavelet Transform
            coeffs = pywt.wavedec(channel_data, wavelet, level=4)
            for j, (band_name, _) in enumerate(bands.items()):
                # Extracting approximation and detail coefficients
                coeff = coeffs[j]
                # Calculating mean and variance
                mean_coeff = np.mean(coeff)
                var_coeff = np.var(coeff)
                feature_vector.extend([mean_coeff, var_coeff])
        
        features[i, :] = feature_vector
    
    return features

# Extracting time-frequency features
tf_features = extract_time_frequency_features(eeg_samples)

# Splitting the dataset into training and testing sets
X_train_tf, X_test_tf, y_train_tf, y_test_tf = train_test_split(tf_features, labels, test_size=0.2, random_state=42)

# Feature scaling
scaler_tf = StandardScaler()
X_train_tf = scaler_tf.fit_transform(X_train_tf)
X_test_tf = scaler_tf.transform(X_test_tf)

X_train_tf.shape, X_test_tf.shape, y_train_tf.shape, y_test_tf.shape
```

```python
from sklearn.ensemble import RandomForestClassifier

# Training a Random Forest classifier
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train_tf, y_train_tf)

# Predicting on the test set with the Random Forest model
y_pred_rf = rf_model.predict(X_test_tf)

# Evaluating the Random Forest model
rf_accuracy = accuracy_score(y_test_tf, y_pred_rf)
rf_classification_report = classification_report(y_test_tf, y_pred_rf, target_names=['Alert', 'Drowsy'])

rf_accuracy, rf_classification_report
```

```python
# Function to calculate spectral features: spectral entropy and peak frequency
def spectral_features(eeg_sample, sample_rate):
    # Spectral Entropy
    freqs, psd = welch(eeg_sample, fs=sample_rate, nperseg=384)
    psd_norm = psd / psd.sum(axis=1, keepdims=True)
    spectral_entropy = -np.sum(psd_norm * np.log2(psd_norm), axis=1)
    
    # Peak Frequency
    peak_freq = freqs[np.argmax(psd, axis=1)]
    
    return np.concatenate([spectral_entropy, peak_freq])

# Function to calculate statistical features: skewness and kurtosis
def statistical_features(eeg_sample):
    skewness = skew(eeg_sample, axis=1)
    kurt = kurtosis(eeg_sample, axis=1)
    return np.concatenate([skewness, kurt])

# Function to calculate Hjorth parameters
def hjorth_parameters(eeg_sample):
    activity = np.var(eeg_sample, axis=1)
    mobility = np.sqrt(np.var(np.diff(eeg_sample, axis=1), axis=1) / activity)
    complexity = np.sqrt(np.var(np.diff(np.diff(eeg_sample, axis=1), axis=1), axis=1) / np.var(np.diff(eeg_sample, axis=1), axis=1))
    return np.concatenate([activity, mobility, complexity])

# Extracting the advanced features for each EEG sample
streamlined_features = np.array([np.concatenate([spectral_features(sample, sample_rate), 
                                                 statistical_features(sample), 
                                                 hjorth_parameters(sample)]) for sample in eeg_samples])

# Splitting the dataset into training and testing sets with streamlined features
X_train_streamlined, X_test_streamlined, y_train_streamlined, y_test_streamlined = train_test_split(
    streamlined_features, labels, test_size=0.2, random_state=42)

# Feature scaling for streamlined features
scaler_streamlined = StandardScaler()
X_train_streamlined = scaler_streamlined.fit_transform(X_train_streamlined)
X_test_streamlined = scaler_streamlined.transform(X_test_streamlined)

X_train_streamlined.shape, X_test_streamlined.shape
```
